```python
import pandas as pd
import sys
sys.path.append('Tanimoto_similarity')
from similarities import tanimoto_similarity


# 1. Load data
smi = pd.read_csv('Tanimoto_similarity/test_smiles.smi', header = None, sep="\t")
smiles_list=smi.iloc[:,0].tolist() #convert dataframe to smiles list

# 2. Call class tanimoto_similarity
simi = tanimoto_similarity(smiles_list = smiles_list, fps ='ecfp2')

# 3. Visualize triangle heatmap similarity matrix
triangle_heatmap = simi.visualize_triangle()
display(triangle_heatmap)

# 4. visualize Square heatmap similarity matrix
square_heatmap = simi.visualize_square()
display(square_heatmap)

# 5. visualize two molecules having best similarity
pair_visualize = simi.pair_best_similarity()
display(pair_visualize)
```
